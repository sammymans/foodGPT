### Bar Chart for Number of Ratings per User

```python
# Create a bar chart with Number of Ratings per User
fig, ax = plt.subplots()
fig.set_facecolor('#F3F3FC')

ratings_per_user = combined.groupby('user_id')['rating'].count()

ax.bar(ratings_per_user.index, ratings_per_user.values)
ax.set_xlabel('User ID')
ax.set_ylabel('Number of Ratings')
ax.set_title('Number of Ratings per User')
ax.set_ylim(0, 400)

plt.show()
```
